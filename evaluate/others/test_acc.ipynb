## Fixed insertion-deletion rates

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have a DataFrame 'df' with columns "Quality" and "Accuracy"
trend_df = df

# Calculate the trend line using NumPy polyfit
trend_coefficients = np.polyfit(range(len(trend_df["Quality"])), trend_df["Accuracy"], deg=1)
trend_line = np.poly1d(trend_coefficients)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(trend_df["Quality"], trend_df["Accuracy"], marker='o', linestyle='-', color="grey", alpha=0.5)
# # Plot the trend line
# plt.plot(trend_line, label="Trend Line", color='red', linewidth=2)
# Set the title and axis labels
plt.title("Fixed IR, DR")
plt.xlabel("Quality Score")
plt.ylabel("Accuracy")
# Add gridlines for better visualization
plt.grid(color='lightgrey', linestyle='--')
# Show the legend
# plt.legend()
# Save the plot as a publication-worthy image
plt.savefig("figures/fixed_ir_dr.png", dpi=300)
# Show the plot
plt.show()
```
